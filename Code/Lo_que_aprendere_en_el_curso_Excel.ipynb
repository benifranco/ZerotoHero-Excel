Puedes utilizar la biblioteca openpyxl asi como pandas para trabajar con archivos de Excel en Python. openpyxl te permite leer y escribir datos en archivos de Excel, así como realizar operaciones en celdas y hojas. Aquí tienes un ejemplo de cómo leer y escribir datos en un archivo de Excel, y luego aplicar un formato condicional.

Instala la biblioteca openpyxl si aún no lo has hecho, para ello utiliza la instrucción
!pip install openpyxl


````python
import openpyxl
from openpyxl.styles import PatternFill
from openpyxl.formatting.rule import CellIsRule

# Datos de ejemplo
data = [
    ["Producto", "Ventas"],
    ["A", 50],
    ["B", 100],
    ["C", 75],
    ["D", 200],
    ["E", 150]
]

# Crear un nuevo archivo de Excel y agregar datos
workbook = openpyxl.Workbook()
sheet = workbook.active

for row in data:
    sheet.append(row)

# Aplicar formato condicional para resaltar ventas superiores a 100
yellow_fill = PatternFill(start_color="FFFF00", end_color="FFFF00", fill_type="solid")
rule = CellIsRule(operator="greaterThan", formula=['100'], stopIfTrue=True, fill=yellow_fill)
sheet.conditional_formatting.add("B2:B6", rule)

# Guardar el archivo de Excel
workbook.save("ventas.xlsx")
````

````python
import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 4]
y = [2, 4, 6, 8, 10]

# Estilo de línea y marcador
plt.plot(x, y, linestyle='--', marker='o', markersize=8, markerfacecolor='red', markeredgecolor='black', linewidth=2)

# Etiquetas de los ejes y título
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.title("Gráfica de ejemplo mejorada")

# Establecer límites para los ejes
plt.xlim(-0.5, 4.5)
plt.ylim(0, 12)

# Agregar cuadrícula
plt.grid(True, linestyle='--', alpha=0.6)

# Personalizar colores y estilos de las marcas en los ejes
plt.tick_params(axis='both', colors='blue', direction='out', length=6, width=2, labelsize=12)

# Agregar anotaciones
for i, v in enumerate(y):
    plt.text(i - 0.15, v + 0.5, str(v), color='blue', fontsize=12)

# Mostrar la gráfica
plt.show()
````

## Probabilidad con Python: 

Puedes utilizar Python para calcular probabilidades y realizar simulaciones de eventos aleatorios. La biblioteca random es útil para generar números aleatorios y simular eventos

````python
import random

lanzamientos = 10000
caras = 0

for _ in range(lanzamientos):
    moneda = random.randint(0, 1)
    if moneda == 1:
        caras += 1

probabilidad_caras = caras / lanzamientos
print(f"Probabilidad de obtener cara: {probabilidad_caras}")

#Este código simula 10,000 lanzamientos de moneda y calcula la probabilidad de obtener cara.
````

##Regresión lineal y logistica con Python

Imagina que eres un chef en un restaurante, y te das cuenta de que la cantidad de felicidad que sienten tus clientes depende de la cantidad de queso que pones en sus hamburguesas. ¡Más queso, más felicidad! Entonces, decides investigar si hay alguna forma de predecir cuánta felicidad tendrán tus clientes según la cantidad de queso que añades a sus hamburguesas.

Para hacer esto, recopilas datos de tus clientes y mides la cantidad de queso en sus hamburguesas (en onzas) y la felicidad que expresan (en una escala de 1 a 10). La regresión lineal es como una receta mágica que te ayuda a encontrar la "mejor línea recta" que describe la relación entre el queso y la felicidad.

La "mejor línea recta" es como un tobogán, donde la cantidad de queso es la altura desde donde empiezas a deslizarte, y la felicidad es qué tan lejos llegas al deslizarte. La receta mágica de regresión lineal te ayuda a ajustar la inclinación y la posición del tobogán para que encaje lo mejor posible con los datos que has recopilado.

Una vez que tengas la mejor línea recta, puedes predecir cuánta felicidad tendrán tus clientes cuando agregues cierta cantidad de queso a sus hamburguesas. Así, podrás crear hamburguesas perfectas y mantener a tus clientes llenos de felicidad.

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Parámetros de la regresión lineal
b = 2.0
alpha = 1.5
mu = 0
sigma = 1.0
n = 100

# Generar datos con la fórmula y = b + alpha * x + e
np.random.seed(42)
x = np.random.rand(n) * 10  #FELICIDAD
e = np.random.normal(mu, sigma, n)
y = b + alpha * x + e   #queso usado en la hamburguesa

# Ajustar el modelo de regresión lineal
X = x.reshape(-1, 1)
modelo = LinearRegression()
modelo.fit(X, y)

# Predecir y trazar la línea de regresión
y_pred = modelo.predict(X)
plt.scatter(x, y, label="Datos") #felicidad vs queso
plt.plot(x, y_pred, color="red", label="Regresión lineal")
plt.xlabel("Felcidad ")
plt.ylabel("Queso")
plt.legend()
plt.show()

print(f"Intersección estimada (b): {modelo.intercept_:.2f}")
print(f"Coeficiente estimado (alpha): {modelo.coef_[0]:.2f}")
````
